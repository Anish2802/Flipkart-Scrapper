```python
from bs4 import BeautifulSoup as bs
import requests as rt
```

```python
flipkart="https://www.flipkart.com/search?q="
```

```python
prod_name=input("Enter the product name")
prod_name
```

```python
prod_link=flipkart+prod_name
```

```python
prod_link
```

```python
web=rt.get(prod_link)
```

```python
web.text
```

```python
data=bs(web.text,"lxml")
```

```python
data
```

```python
try:
    store_prod_name=data.find('div',class_="_4rR01T").text
    store_prod_price=data.find('div',class_="_30jeq3 _1_WHN1").text
    store_prod_link="https://www.flipkart.com"+data.find('a',class_="_1fQZEK").get("href")
except Exception:
    store_prod_name=data.find('a',class_="IRpwTa").text
    store_prod_price=data.find('div',class_="_30jeq3").text
    store_prod_link="https://www.flipkart.com"+data.find('a',class_="IRpwTa").get("href")    
```

```python
print("NAME:",store_prod_name,"\n","PRICE:",store_prod_price,"\n","LINK:",store_prod_link)
```
